```python
def _nameAndRole(personList, joiner=u', '):
    """Build a pretty string with name and role."""
    nl = []
    for person in personList:
        n = person.get('name', u'')
        if person.currentRole: n += u' (%s)' % person.currentRole
        nl.append(n)
    return joiner.join(nl)
```
